```python
import mlflow
from mlflow.tracking import MlflowClient


def get_tags(run_id):

    client = MlflowClient()

    run = client.get_run(run_id)
    tags = run.data.tags

    keys = ['label','sampling','type','strategy','date_interval','date_filter']

    select_tags = {}
    for k in keys:
        select_tags[k] = tags[k]

    select_tags['n_days'] = int(tags['label'].replace("churn",""))

    return select_tags


def filter_data(run_id, processed_date):
    
    tags = get_tags(run_id)

    # Processed_date - 

    # Processed date should be when the prediction was made based on the date filter - number of days back 
    preds = spark.table(PREDICTION_TABLE_NAME).filter(col('label_name')==tags['label']).filter(col('processed_date')==processed_date).filter(col('model_run_id')==run_id)

    # Processed date should be when the prediction was made based on the date filter - number of days back
    labels = spark.table(LABEL_TABLE_NAME).filter(col('label_name')==tags['label']).filter(col('processed_date')== date_sub(processed_date,tags['n_days']))

    return preds, labels


```
